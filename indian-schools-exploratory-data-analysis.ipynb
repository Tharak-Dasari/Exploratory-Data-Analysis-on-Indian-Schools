# What is an EDA Project?

Exploratory Data Analysis (EDA) is a critical process that helps data analysts to gain insights into the data and identify patterns, anomalies, and trends. It involves examining and visualizing the data to understand its distribution, relationships, and statistical properties. EDA is typically the first step in any data analysis project, and it helps to guide the rest of the analysis.

EDA involves a range of techniques, including descriptive statistics, data visualization, and data transformation. By using these techniques, data analysts can identify data quality issues, detect outliers and anomalies, and gain a better understanding of the underlying patterns in the data. This process can help analysts to identify potential problems or opportunities, and to formulate hypotheses that can be tested using more advanced analytical methods.

Overall, EDA is a crucial step in any data analysis project, and it helps to ensure that analysts have a good understanding of the data before they begin more complex analyses. To learn more about EDA, you can refer to this [link](https://en.wikipedia.org/wiki/Exploratory_data_analysis) to the Wikipedia page on Exploratory Data Analysis.

# Methodology Employed 
It is common for a typical EDA project to involve certain sequential steps in order to conduct a thorough analysis and gain a better understanding of the data.They are:

*   Data collection from various sources and formats (e.g. CSV, Excel).
*   Data cleaning to remove inconsistencies and errors.
*   Data transformation to prepare the data for analysis.
*   Summarization of data using descriptive statistics (e.g. mean, median, standard deviation).
*   Visualization of data using techniques such as barcharts, scatter plots, and pie-charts.
*   Analysis of insights gained from EDA to inform further analyses and modeling.
*   Asking and answering interesting questions related to the data to uncover hidden insights and potential correlations.
*   Summarizing the inferences drawn from the analysis drawing appropriate conclusions based on the results obtained.

#### 1.Removing Impractical or Non-Genuine Data
In the process of data analysis, it is often necessary to eliminate rows that contain values that are deemed impractical or erroneous. By removing such data, we can focus on working with feasible and practical information, thus ensuring the reliability and accuracy of our analysis.

To drop rows based on specific conditions, we can use the [.drop()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html) method in pandas. This method allows us to remove rows that do not meet certain criteria, such as values above or below a practical threshold. The general common sense and domain knowledge play a crucial role in determining these practical thresholds.






```python
#let us make a duplicate dataframe
Schools_data = Dataframe.copy()
```

```python
# let us drop some values which we think are not genuine for analysis

Schools_data.drop(Schools_data[Schools_data.Establishment > 2022].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data.Establishment < 1900].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data['Male Teachers'] > 50].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data['Female Teacher'] > 50].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data['Boys Toilet'] > 50].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data['Girls Toilet'] > 50].index, inplace=True)

Schools_data.drop(Schools_data[Schools_data['Computers'] > 250].index, inplace=True)
```

After careful consideration and analysis of the data, I have made the decision to drop certain rows based on my understanding of the threshold values and distributions of specific data. This step was taken to remove outliers from the dataset, even if they were deemed genuine, in order to improve the quality and visual representation of the data during visualization. By eliminating outliers, the overall visualizations are less likely to be skewed or heavily influenced by a few extreme values, allowing for a clearer and more accurate representation of the data patterns and trends.

#### 2.Replacing null values
In data analysis, handling null or missing values is a crucial step to ensure the integrity and completeness of the dataset. Null values can occur due to various reasons, such as data collection errors or incomplete records. Replacing null values with appropriate alternatives is necessary to avoid biased or incorrect analyses.

In order to handle null values in the columns 'Instruction Medium', 'Building', 'School Area', 'Library', 'Management', and 'School Type', I have utilized the [.fillna()](https://www.w3schools.com/python/pandas/ref_df_fillna.asp) method. Specifically, I have chosen to fill the null values with the previous non-null value from the same column using the [forward fill (ffill)](https://www.w3schools.com/python/pandas/ref_df_ffill.asp) method. This approach assumes that the other values within the same state remain the same where the null values are present. By employing this method, I aim to maintain consistency within the dataset and mitigate the impact of missing values on subsequent analysis.

```python
# let us replace null values with approriate values.

Schools_data['Instruction Medium'].fillna(method = 'ffill', inplace=True)
Schools_data['Building'].fillna(method = 'ffill', inplace=True)
Schools_data['School Area'].fillna(method = 'ffill', inplace=True)
Schools_data['Library'].fillna(method = 'ffill', inplace=True)
Schools_data['Management'].fillna(method = 'ffill', inplace=True)
Schools_data['School Type'].fillna(method = 'ffill', inplace=True)
```

### 1.States with heighst and lowest number of schools.

### **Q1.What is the gender distribution among teachers in Indian Schools?**

Gender distribution of teachers in schools in India is an important aspect to analyze as it provides insights into the representation and balance of male and female teachers in the education system. Understanding the gender distribution can help identify any potential gender disparities or imbalances that may exist in the teaching workforce. This analysis can shed light on factors such as gender equity, equal opportunities, and the overall inclusivity of the education system.

Let us use [Histogram](https://seaborn.pydata.org/generated/seaborn.histplot.html) to plot this analysis. The plot includes the percentage distribution of genders in teachers in schools and the no.of schools.

```python
# Calculate the percentage of male and female teachers in each school
schools_df['total_teachers'] = schools_df['Male Teachers'] + schools_df['Female Teacher']
schools_df['male_percentage'] = (schools_df['Male Teachers'] / schools_df['total_teachers']) * 100
schools_df['female_percentage'] = (schools_df['Female Teacher'] / schools_df['total_teachers']) * 100

# Create a histogram to visualize the distribution of gender ratios in schools
plt.figure(figsize=(8,3))
sns.histplot(data=schools_df, x='male_percentage', kde=False, bins=30, color='blue', alpha=0.5, label='Male Teachers')
sns.histplot(data=schools_df, x='female_percentage', kde=False, bins=30, color='pink', alpha=0.5, label='Female Teachers')

# Customize the graph
plt.xlabel("Percentage of Teachers")
plt.ylabel("Number of Schools")
plt.title("Gender Distribution of Teachers in Schools",fontsize=20)
plt.legend()
plt.show()
```

This graph demonstrates the gender distribution of teachers in schools.This histogram can be a **bit confusing but quite interesting** to read because both the male and female distributions are plotted on top of each other. However, it allows us to compare the distribution of male and female teachers in schools and see if there are any noticeable patterns or differences.

We can see that there are comparitively more schools where percentage of male teachers is 100% and a clear trend can be seen where the number of schools are increasing with the increase of percentage in male teachers.(Compared to the no.of schools with same percentage of female teachers)

It is also worth noting that there are some schools where the percentage of male teachers is much higher than the percentage of female teachers, while in others it is the opposite. This suggests that there may be regional or cultural factors that influence the gender distribution of teachers in different parts of the country.


Let us also see the overall distribution of genders among teachers all over India. Let us use a Pie-Chart to show this comparison.

```python
# Calculate the total count of male and female teachers
total_male_teachers = schools_df['Male Teachers'].sum()
total_female_teachers = schools_df['Female Teacher'].sum()

# Create a list of the values for the pie chart
pie_values = [total_male_teachers, total_female_teachers]

# Create a list of the labels for the pie chart
pie_labels = ['Male Teachers', 'Female Teachers']

# Create the pie chart
plt.pie(pie_values, labels=pie_labels, autopct='%1.1f%%', startangle=90, radius=1)

# Add a title to the chart
plt.title("Distribution of Total Male and Female Teachers in Indian Schools",fontsize=12)

# Show the chart
plt.show()
```

The pie Chart shows that the overall distribution of male and female teachers in Indian schools is almost evenly split, with male teachers making up around 52% and female teachers comprising around 48% of the total number of teachers. This indicates that there has been a significant increase in the number of female teachers in Indian schools in recent years, reflecting the changing social and economic landscape of the country.

We will create a chloropleth map using JSON and Folium tools, along with a JavaScript function to handle hover properties. This approach allows us to leverage the power of these tools to generate an interactive map. By doing so, we can visualize the data in a meaningful way and enable users to explore and interact with the map to gain insights about the regions represented in the chloropleth map.

```python
#Import required libraries and tools
import requests
import json
import folium

# Download the GeoJSON data and save it locally
url = 'https://raw.githubusercontent.com/Subhash9325/GeoJson-Data-of-Indian-States/master/Indian_States'
r = requests.get(url)
with open('indian_states.geojson', 'w') as f:
    f.write(r.text)

# Prepare the data
state_counts = schools_df.groupby('State')['School_Name'].count().reset_index()
state_counts.columns = ['State', 'Count']

# Group the data by state and school area
area_counts = schools_df.groupby(['State', 'School Area'])['School_Name'].count().unstack('School Area', fill_value=0).reset_index()
area_counts.columns = ['State', 'Urban', 'Rural']

# Load the GeoJSON data
with open('indian_states.geojson') as f:
    geojson = json.load(f)

# Merge the data
merged_data = geojson.copy()
for i in range(len(merged_data['features'])):
    state_name = merged_data['features'][i]['properties']['NAME_1']
    count = state_counts[state_counts['State'] == state_name]['Count'].values
    urban_count = area_counts[area_counts['State'] == state_name]['Urban'].values
    rural_count = area_counts[area_counts['State'] == state_name]['Rural'].values
    if len(count) > 0:
        merged_data['features'][i]['properties']['Count'] = int(count)
        merged_data['features'][i]['properties']['Urban'] = int(urban_count)
        merged_data['features'][i]['properties']['Rural'] = int(rural_count)
    else:
        merged_data['features'][i]['properties']['Count'] = -1
        merged_data['features'][i]['properties']['Urban'] = -1
        merged_data['features'][i]['properties']['Rural'] = -1

# Create the choropleth map
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Custom JavaScript function to display total schools count on hover
hover_code = """
function(e) {
    var layer = e.target;
    layer.bindTooltip("<b>State: </b>" + layer.feature.properties.NAME_1 + "<br><b>Total Schools: </b>" + layer.feature.properties.Count, {sticky: true});
    layer.openTooltip();
}
"""

folium.Choropleth(
    geo_data=merged_data,
    name='choropleth',
    data=state_counts,
    columns=['State', 'Count'],
    key_on='feature.properties.NAME_1',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    nan_fill_color='white',
    legend_name='Number of Schools',
    highlight=True,
).add_to(m)

# Add hover functionality using custom JavaScript
folium.GeoJson(
    merged_data,
    name='geojson',
    style_function=lambda feature: {
        'fillOpacity': 0,
        'color': 'white',
        'weight': 1
    },
    highlight_function=lambda feature: {
        'fillOpacity': 0.7,
        'color': 'gray',
        'weight': 1
    },
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1', 'Count'],
        aliases=['State', 'Total Schools'],
        style="font-size: 10px",
        localize=True
    ),
    smooth_factor=0
).add_to(m)

# Add the custom JavaScript to the map
m.get_root().html.add_child(folium.Element('<script>' + hover_code + '</script>'))

# Add layer control
folium.LayerControl().add_to(m)

# Show the map
m
```

The code generates a choropleth map of India, displaying the number of schools in each state, when hovered over the particular state. It downloads the GeoJSON data and merges it with the school dataset, associating school counts with each state. The resulting map provides an interactive visualization where darker shades indicate higher school counts. Hovering over a state displays a tooltip with the state name and the total number of schools. This interactive map allows users to explore school distributions and gain insights into the educational landscape across different regions of India.

Conversely, states like Rajasthan,Madhya Pradesh and Maharastra have lower school densities, possibly due to their larger land areas and lower concentration of schools in proportion to their size.

Let's see the density of schools for each state clearly for better understanding.

```python
#import required libraries and tools
import seaborn as sns
import matplotlib.pyplot as plt

# Group the schools by state and calculate the average state area and total number of schools
grouped_df = merged_df.groupby('State').agg({'State_Area in sq.KM': 'mean', 'UDISE Code ': 'count'}).reset_index()
grouped_df.rename(columns={'UDISE Code ': 'School Count'}, inplace=True)

# Calculate the density of schools per sq. km for each state
grouped_df['School Density'] = grouped_df['School Count'] / grouped_df['State_Area in sq.KM']

# Sort the dataframe by school density in descending order
grouped_df = grouped_df.sort_values(by='School Density', ascending=False)

# Create a bar chart with the state name on the x-axis and school density on the y-axis
plt.figure(figsize=(12,3))
sns.barplot(x='State', y='School Density', data=grouped_df)

# Set the title and x/y-axis labels
plt.title('Density of Schools by State', fontsize=16)
plt.xlabel('State', fontsize=10)
plt.ylabel('School Density (per sq. km)', fontsize=10)

# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

so, now literacy rate (in %)  of the states is added next to its geographical area. let us visualize the corelation between the school density of states and their literacy rate using a scatter plot which is interactive.

```python
import plotly.express as px

# Group schools by state and calculate the density of schools
schools_by_state = merged_df.groupby('State')['School_Name'].count()  # Count the number of schools in each state
state_area = merged_df.groupby('State')['State_Area in sq.KM'].mean()  # Calculate the average area of each state
school_density = schools_by_state / state_area  # Calculate the school density for each state

# Create a scatter plot to visualize the correlation between school density and literacy rate
fig = px.scatter(merged_df, x='Literacy_Rate (%)', y=school_density[merged_df['State'].values],
                 hover_data=['State'], title='Correlation between School Density and Literacy Rate')

# Customize the scatter plot
fig.update_traces(marker=dict(size=10, opacity=0.3, color='orange', line=dict(width=1, color='Blue')))
fig.update_layout(title=dict(x=0.5, y=0.98, xanchor='center', yanchor='top'),
                  xaxis_title='Literacy Rate (%)', yaxis_title='School Density (schools/sq.km)')

# Display the scatter plot
fig.show()
```

*We then use Plotly Express to create a scatter plot. The x-axis represents the literacy rate (%) of each state, and the y-axis represents the school density. Each data point on the scatter plot represents a state, with the size, opacity, and color of the marker customized for better visualization.*

Upon analyzing the scatter plot, it is evident that there is no significant correlation between the literacy rate and the density of schools in most states, except for Union Territories (UTs). This lack of correlation suggests that factors other than school density may influence the literacy rates in states.

In states with low school density but relatively high literacy rates, alternative educational systems, such as private institutions or distance learning, could be prevalent. Conversely, states with high school density but lower literacy rates might face challenges related to quality of education, teacher-student ratios, or infrastructure limitations.

It is crucial to consider the broader context and socioeconomic aspects while interpreting the relationship between school density and literacy rates.
